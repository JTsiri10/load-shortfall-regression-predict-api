```python
#After we have cleaned our data with some data engineering we use the OLS model to detect multicollinierity

#importing old model
from statsmodels.formula.api import ols
#Model DataFrame with all of the columns:
dfm = df_clean.copy() #avoiding losing our original data

# The dependent variable:
y_name = 'load_shortfall_3h'
# The independent variable
# (let's first try all of the columns in the model DataFrame)
X_names = [col for col in dfm.columns if col != y_name]

# Build the OLS formula string " y ~ X "
formula_str = y_name+" ~ "+" + ".join(X_names);
print('Formula:\n\t {}'.format(formula_str))
```

```python
#fitting ols model
model=ols(formula=formula_str, data=dfm)
fitted = model.fit()

# Output the fitted summary
print(fitted.summary())
```

```python
#From the OLS model we see that the Durbin-Watson 0.4 and the r Squared is 0.14
#This tells us that thereis multicolinierity and the current features 
#we have are not effectively explaining our target variable
```

```python
#checking highly positive correlated variables so that we can drop some
#This function loops through the whole correlation matrix to check if correlation is above threshold

def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
#checking data that fits our threshold
corr_features=correlation(df,0.8)
len(set(corr_features))

#display the list of features with mutlicolinierity
corr_features
```

```python
# Lets take a new subset of our potential independent variables
#drop colomns that have mutlicolinierity
X_remove = ['Barcelona_temp_min','Barcelona_temp_max','Bilbao_temp_max','Madrid_temp_min','Madrid_temp_max','Seville_temp_min','Valencia_temp_min']
X_corr_names = [col for col in X_names if col not in X_remove]

# Create our new OLS formula based-upon our smaller subset
formula_str = y_name+' ~ '+' + '.join(X_corr_names);
print('Formula:\n\t{}'.format(formula_str))
```

```python
# Fit the OLS model using the model dataframe
model=ols(formula=formula_str, data=dfm)
fitted = model.fit()

# Display the fitted summary
print(fitted.summary())
```

```python
## Bring all imports in notebook to here with comments to indicate the different 
## accuracy packages
from sklearn.metrics import mean_squared_error
from sklearn import *
from sklearn.linear_model import *
```

```python
# create targets and features dataset

y = df_clean[:len(df_train)][['load_shortfall_3h']]

x = df_clean[:len(df_train)].drop('load_shortfall_3h',axis=1)

#To be used later for submission to Kaggle
x_train = df_clean[:len(df_train)].drop('load_shortfall_3h',axis=1)
x_test = df_clean[len(df_train):].drop('load_shortfall_3h',axis=1)
```

```python
#view x
x.head()
```

```python
#view y
y.head()
```

```python
# split data into validation and training data
# Using just the training dataset to test model accuracy 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2) 
```

```python
# Get training and testing data for correlation threshold model
X_corr_train = x_train[X_corr_names]
X_corr_test = x_test[X_corr_names]
```

```python
# create one or more ML models
```

```python
#Importing libraries for models
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

#Training the 3 different models
RF=RandomForestRegressor()
RF.fit(X_corr_train,y_train)

LM = LinearRegression()
LM.fit(X_corr_train,y_train)

DT=DecisionTreeRegressor()
DT.fit(X_corr_train,y_train)

```

```python
# evaluate one or more ML models

#Predicting using the models
ModelRF = RF.predict(X_corr_test)
ModelLM = LM.predict(X_corr_test)
ModelDT = DT.predict(X_corr_test)
```

```python
# Compare model performance
# Dictionary of results
results_dict = {'Model MSE':
                    {
                        "Random Forest": np.sqrt(metrics.mean_squared_error(y_test, ModelRF)),
                        "Linear Regression": np.sqrt(metrics.mean_squared_error(y_test, ModelLM)),
                        "Decision Trees": np.sqrt(metrics.mean_squared_error(y_test, ModelDT))
                    },
                    'Model R2':
                    {
                        "Random Forest": format(metrics.r2_score(y_test,ModelRF)),
                        "Linear Regression": format(metrics.r2_score(y_test, ModelLM)),
                        "Decision Trees": format(metrics.r2_score(y_test, ModelDT))
                    }
                }
```

```python
# Choose best model and motivate why it is the best choice

#Graphically present models
# Create figure and axes
f, ax = plt.subplots(figsize=(15,5), nrows=1, ncols=3, sharey=True)

# Create list of titles and predictions to use in for loop
pred = [LM.predict(X_corr_test), DT.predict(X_corr_test), RF.predict(X_corr_test)]
title = ['Linear Regression','Decision tree', 'Random Forest']

# Loop through all axes to plot each model's results 
for i in range(3):
    rmse = round(np.sqrt(mean_squared_error(pred[i],y_test)))
    ax[i].set_title(title[i]+"  (RMSE: "+str(rmse)+ ")")
    ax[i].set_xlabel('Actual')
    ax[i].set_ylabel('Predicted')
    ax[i].plot(y_test,y_test,'r')
    ax[i].scatter(y_test,pred[i])
```
